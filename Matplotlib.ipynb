```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating the dataset on which graphs will be plotted
x = np.linspace(0,10,100) # Give 100 values between 0 and 10
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
# With labels, plotting the cosine graph
plt.plot(x,z)
plt.xlabel('Angles')
plt.ylabel('Cosine values')
plt.title('Cosine graph')
plt.show()
```

```python
# Parabola
x = np.linspace(-10,10,20)
y = x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'g+')
plt.show()
```

```python
x = np.linspace(-10,10,20)
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'r-')
plt.show()
```

# Bar Graph

```python
fig = plt.figure() # Create an empty figure object
ax = fig.add_axes([0,0,1,1]) #[left, bottom, width, height]
langs = ['C','C++','Java','Python','PHP']
students = [23,17,35,29,12]
ax.bar(langs,students)
plt.xlabel('Programming languages')
plt.ylabel('Number of students enrolled')
plt.title('Bar graph')
plt.show()
```

# Pie chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
langs = ['C','C++','Java','Python','PHP']
students = [23,17,35,29,12]
ax.pie(students,labels=langs,autopct='%1.1f%%')
plt.show()
```

# Scatter plot

```python
x = np.linspace(0,10,40)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

#3D scatter plot

```python
fig2 = plt.figure()
ax = plt.axes(projection = '3d')
# x = 20*np.random.random(100)
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
ax.scatter(x,y,x,c=z,cmap='Greens')
plt.show()
```
