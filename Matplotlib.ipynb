```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating the dataset on which graphs will be plotted
x = np.linspace(0,10,100) # Give 100 values between 0 and 10
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
# With labels, plotting the cosine graph
plt.plot(x,z)
plt.xlabel('Angles')
plt.ylabel('Cosine values')
plt.title('Cosine graph')
plt.show()
```

```python
# Parabola
x = np.linspace(-10,10,20)
y = x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'g+')
plt.show()
```
