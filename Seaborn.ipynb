## Seaborn linrary: Data visualization lib

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## seaborn library has some builtin dataset

```python
# Dataset: total bill vs tip
tips = sns.load_dataset('tips')

```

```python
tips.head()
```



*   hue parameter is used to differentiate data points by color
*  style parameter is used to differentiate data points using different
marker styles


*  col='time' -> Splits the plot into separate subplots (facets) based on the time column (Lunch or Dinner).



```python
# Let's visualize the data
# lets set the theme first for better visualization
sns.set_theme()
sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker',style='smoker',size='size',col='time')
plt.show()
```
