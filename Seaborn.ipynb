## Seaborn linrary: Data visualization lib

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## seaborn library has some builtin dataset

```python
# Dataset: total bill vs tip
tips = sns.load_dataset('tips')

```

```python
tips.head()
```



*   hue parameter is used to differentiate data points by color
*  style parameter is used to differentiate data points using different
marker styles


*  col='time' -> Splits the plot into separate subplots (facets) based on the time column (Lunch or Dinner).



```python
# Let's visualize the data
# lets set the theme first for better visualization
sns.set_theme()
sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker',style='smoker',size='size',col='time')
plt.show()
```

```python
from sklearn.datasets import fetch_california_housing
california_dataset = fetch_california_housing()
print(california_dataset)
```

```python
california_df = pd.DataFrame(california_dataset.data, columns = california_dataset.feature_names)
california_df['PRICE'] = california_dataset.target;
```

```python
california_df.head()
```

Distribution plot

```python
sns.distplot(california_df['PRICE'])
plt.show()
```

Correlation

```python
cor = california_df.corr()
```

```python
print(cor)
```

Now constructing a heat map

```python
plt.figure(figsize=(10,10)) # figsize set the area of the graph
sns.heatmap(cor,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
plt.show()
```
