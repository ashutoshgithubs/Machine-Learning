```python
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
  text_clean = text_clean = re.sub(r'[^a-zA-Z\s]', "", messages["message"][i])
  lower_text = text_clean.lower()
  tolenized_text = word_tokenize(lower_text)
  stemmed_word = []
  for word in tolenized_text:
    if not word in stopwords.words('english'):
      words = ps.stem(word)
      stemmed_word.append(words)

  review = ' '.join(stemmed_word)
  corpus.append(review)


```

## AvgWord2Vec
 ##### Here Lemmetized word will be used

```python
!pip install gensim
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
nltk.download('wordnet')
```

```python
corpus2=[]
for i in range(0,len(messages)):
  text_clean = text_clean = re.sub(r'[^a-zA-Z\s]', "", messages["message"][i])
  lower_text = text_clean.lower()
  tolenized_text = word_tokenize(lower_text)
  lemmetized_word = []
  for word in tolenized_text:
    if not word in stopwords.words('english'):
      words = lemmatizer.lemmatize(word)
      lemmetized_word.append(words)
  review = ' '.join(lemmetized_word)
  corpus2.append(review)
```

```python
corpus2
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess # to lowercase the word
```

```python
words=[]
for sent in corpus2:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words
```

### Let's train Word2Vec from Scratch

```python
import gensim
```

```python
model = gensim.models.Word2Vec(words, window=5, min_count=2)
```

```python
model.wv.index_to_key # Gives all vocabularies
```

```python
# Vocabulary size
model.corpus_count
```

```python
model.epochs
```

```python
model.wv.similar_by_word('list')
```

### By default, this word2vec model has 100 sized dimension

### AvgWord2Vec

```python
import numpy as np
```

```python
def avg_word2vec(doc):
  return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key], axis=0)
```

```python
!pip install tqdm
```

```python
from tqdm import tqdm
```

```python
X=[]
for i in tqdm(range(len(words))):
  X.append(avg_word2vec(words[i]))
```
